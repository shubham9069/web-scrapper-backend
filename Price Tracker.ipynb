```python
soup=BeautifulSoup(response.content,"lxml")
```

```python
walmart_obj = {}
```

```python
try:
    walmart_obj["price"] = soup.find("span",{"itemprop":"price"}).text.replace("Now ","")
except:
    walmart_obj["price"]=None
    
try:
    walmart_obj["name"] = soup.find("h1",{"itemprop":"name"}).text
except:
    walmart_obj["name"]=None
    
try:
    walmart_obj["rating"] = soup.find("span",{"class":"rating-number"}).text.replace("(","").replace(")","")
except:
    walmart_obj["rating"]=None
```
